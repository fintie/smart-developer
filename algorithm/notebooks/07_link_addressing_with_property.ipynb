```python
import geopandas as gpd
import pandas as pd
from pathlib import Path
```

```python
address_gdf = gpd.read_parquet("../../data/processed/nsw_addressing/addresspoint_all.parquet")
property_gdf = gpd.read_parquet("../../data/processed/nsw_property/property.parquet")
```

```python
print("Address rows:", len(address_gdf))
print("Property rows:", len(property_gdf))

print("Address CRS:", address_gdf.crs)
print("Property CRS:", property_gdf.crs)

print(address_gdf.columns)
print(property_gdf.columns)
```

```python
property_main = property_gdf[property_gdf["propertytype"] == 1].copy()
print(len(property_main))
```

```python
print("Address gurasid null rate:", address_gdf["gurasid"].isna().mean())
print("Property gurasid null rate:", property_main["gurasid"].isna().mean())
```

```python
print("Address duplicated gurasid rate:", address_gdf["gurasid"].duplicated().mean())
print("Property duplicated gurasid rate:", property_main["gurasid"].duplicated().mean())
```

```python
address_ids = set(address_gdf["gurasid"].dropna().unique())
property_ids = set(property_main["gurasid"].dropna().unique())

common_ids = address_ids & property_ids
print("Common gurasid count:", len(common_ids))
print("Address coverage by property:", len(common_ids) / len(address_ids))
print("Property coverage by address:", len(common_ids) / len(property_ids))
```

```python
print("address dtype:", address_gdf["gurasid"].dtype)
print("property dtype:", property_main["gurasid"].dtype)

print(address_gdf["gurasid"].head(20).tolist())
print(property_main["gurasid"].head(20).tolist())
```

```python
address_ids_series = pd.to_numeric(address_gdf["gurasid"], errors="coerce").astype("Int64")
property_ids_series = pd.to_numeric(property_main["gurasid"], errors="coerce").astype("Int64")

address_ids = set(address_ids_series.dropna().astype(int).unique())
property_ids = set(property_ids_series.dropna().astype(int).unique())

common_ids = address_ids & property_ids
print("Common gurasid count after numeric cast:", len(common_ids))
print("Address coverage by property:", len(common_ids) / len(address_ids))
print("Property coverage by address:", len(common_ids) / len(property_ids))
```

```python
print("Address principaladdresssiteoid null rate:", address_gdf["principaladdresssiteoid"].isna().mean())
print("Property principaladdresssiteoid null rate:", property_main["principaladdresssiteoid"].isna().mean())

print("Address addressstringoid null rate:", address_gdf["addressstringoid"].isna().mean())
print("Property addressstringoid null rate:", property_main["addressstringoid"].isna().mean())
```

```python
addr_site_ids = set(
    pd.to_numeric(address_gdf["principaladdresssiteoid"], errors="coerce")
    .dropna()
    .astype(int)
    .unique()
)
prop_site_ids = set(
    pd.to_numeric(property_main["principaladdresssiteoid"], errors="coerce")
    .dropna()
    .astype(int)
    .unique()
)

common_site_ids = addr_site_ids & prop_site_ids
print("Common principaladdresssiteoid count:", len(common_site_ids))
print("Address coverage by principaladdresssiteoid:", len(common_site_ids) / len(addr_site_ids))
print("Property coverage by principaladdresssiteoid:", len(common_site_ids) / len(prop_site_ids))
```

```python
addr_str_ids = set(
    pd.to_numeric(address_gdf["addressstringoid"], errors="coerce")
    .dropna()
    .astype(int)
    .unique()
)
prop_str_ids = set(
    pd.to_numeric(property_main["addressstringoid"], errors="coerce")
    .dropna()
    .astype(int)
    .unique()
)

common_str_ids = addr_str_ids & prop_str_ids
print("Common addressstringoid count:", len(common_str_ids))
print("Address coverage by addressstringoid:", len(common_str_ids) / len(addr_str_ids))
print("Property coverage by addressstringoid:", len(common_str_ids) / len(prop_str_ids))
```

```python
address_attr = address_gdf[
    ["rid", "principaladdresssiteoid", "address", "housenumber", "urbanity"]
].copy()

property_attr = property_main[
    [
        "RID",
        "principaladdresssiteoid",
        "propertytype",
        "valnetpropertystatus",
        "valnetpropertytype",
        "dissolveparcelcount",
        "valnetlotcount",
        "propid",
        "superlot",
        "address",
        "housenumber",
        "urbanity",
        "Shape__Area",
        "Shape__Length",
    ]
].copy()

linked_site = address_attr.merge(
    property_attr,
    on="principaladdresssiteoid",
    how="inner",
    suffixes=("_address", "_property"),
)

print("Linked rows:", len(linked_site))
print("Duplicated rid_address rate:", linked_site["rid"].duplicated().mean())
print("Duplicated RID_property rate:", linked_site["RID"].duplicated().mean())
```
