```python
import pandas as pd
from pathlib import Path
from algorithm.src.scoring.scoring import score_dataframe
```

```python
df = pd.read_parquet("../../data/processed/site_features/property_site_features_v1_sample.parquet")
scored = score_dataframe(df)

print(len(scored))
scored.head()
```

```python
score_cols = [c for c in scored.columns if c.endswith("_score")]
scored[score_cols].describe()
```

```python
base_cols = [
    "RID",
    "address",
    "lot_size_proxy_sqm",
    "primary_zoning_code",
    "primary_zoning_class",
    "mixed_zoning_flag",
    "heritage_flag",
    "heritage_max_significance",
    "bushfire_flag",
    "bushfire_risk_level",
    "flood_flag",
    "primary_flood_class",
    "distance_to_station_m",
    "within_800m_catchment",
]
```

```python
def inspect_strategy(df: pd.DataFrame, score_col: str, n: int = 20):
    print(f"\n=== TOP {n}: {score_col} ===")
    display(
        df.sort_values(score_col, ascending=False)[base_cols + [score_col]].head(n)
    )

    print(f"\n=== BOTTOM {n}: {score_col} ===")
    display(
        df.sort_values(score_col, ascending=True)[base_cols + [score_col]].head(n)
    )
```

```python
inspect_strategy(scored, "single_dwelling_rebuild_score")
```

```python
inspect_strategy(scored, "assembly_opportunity_score")
```

```python
for score_col in score_cols:
    print(f"\n=== {score_col} top 10% zoning distribution ===")
    cutoff = scored[score_col].quantile(0.9)
    top_df = scored[scored[score_col] >= cutoff]
    print(top_df["primary_zoning_code"].value_counts(dropna=False).head(15))
```

```python
for score_col in score_cols:
    print(f"\n=== {score_col} top 10% lot size summary ===")
    cutoff = scored[score_col].quantile(0.9)
    top_df = scored[scored[score_col] >= cutoff]
    print(top_df["lot_size_proxy_sqm"].describe())
```

```python
for score_col in score_cols:
    print(f"\n=== {score_col} top 10% constraints ===")
    cutoff = scored[score_col].quantile(0.9)
    top_df = scored[scored[score_col] >= cutoff]

    print("heritage_flag:")
    print(top_df["heritage_flag"].value_counts(normalize=True, dropna=False))

    print("bushfire_risk_level:")
    print(top_df["bushfire_risk_level"].value_counts(normalize=True, dropna=False))

    print("flood_flag:")
    print(top_df["flood_flag"].value_counts(normalize=True, dropna=False))
```

```python
for score_col in score_cols:
    print(f"\n=== {score_col} top 10% accessibility ===")
    cutoff = scored[score_col].quantile(0.9)
    top_df = scored[scored[score_col] >= cutoff]

    print(top_df["distance_to_station_m"].describe())
    print(top_df["within_800m_catchment"].value_counts(normalize=True, dropna=False))
```

```python
weird_apartment = scored[
    (scored["low_rise_apartment_score"] >= 70)
    & (
        scored["primary_zoning_code"].isin(["R2", "RU1", "RU4", "C2", "C3", "C4", "RE1"])
        | (scored["heritage_flag"] == 1)
        | (scored["flood_flag"] == 1)
    )
]

weird_apartment[base_cols + ["low_rise_apartment_score"]].head(30)
```

```python
missed_apartment = scored[
    (scored["primary_zoning_code"].isin(["MU1", "R4", "SP5", "E1", "E2"]))
    & (scored["lot_size_proxy_sqm"] >= 2000)
    & (scored["distance_to_station_m"] <= 2000)
    & (scored["low_rise_apartment_score"] < 50)
]

missed_apartment[base_cols + ["low_rise_apartment_score"]].head(30)
```

```python
Path("../../data/processed/site_features").mkdir(parents=True, exist_ok=True)

scored.to_parquet(
    "../../data/processed/site_features/property_site_features_v1_sample_scored.parquet",
    index=False
)
```
